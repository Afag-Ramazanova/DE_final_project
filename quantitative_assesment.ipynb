```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = q10.copy()
plt.figure(figsize=(10, 6))
plt.plot(data['Requests/s'], data['Median Response Time'], label='Median Response Time')
plt.plot(data['Requests/s'], data['Average Response Time'], label='Average Response Time')
plt.xlabel('Requests per Second (RPS)')
plt.ylabel('Response Time (ms)')
plt.title('Response Time vs Requests per Second')

plt.xlim(0, data['Requests/s'].max() + 10)  # Add some padding
plt.ylim(0, data[['Median Response Time', 'Average Response Time']].max().max() + 10)

plt.legend()
plt.show()

```

```python

# Filter out only meaningful rows
q10 = q10[q10["Name"] != "Aggregated"]

# Plot Latency vs. RPS
plt.figure(figsize=(10, 6))
plt.plot(q10["Request Count"], q10["Median Response Time"], label="Median Latency")
plt.plot(q10["Request Count"], q10["95%"], label="95th Percentile Latency")
plt.xlabel("Requests per Second")
plt.ylabel("Latency (ms)")
plt.title("Latency vs Requests per Second")
plt.legend()
plt.show()
```
