### Path to datasets: https://www.kaggle.com/datasets/lokeshparab/amazon-products-dataset?select=Air+Conditioners.csv

```python
import os
import pandas as pd

# Initialize an empty list to store DataFrames
dfs = []

# Path to the input directory
input_dir = '/kaggle/input/amazon-products-dataset'

for filename in os.listdir(input_dir):
    if filename.endswith('.csv'): 
        file_path = os.path.join(input_dir, filename)
        # Read each CSV file and append it to the list
        df = pd.read_csv(file_path)
        dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)

# Display the shape and first few rows of the combined DataFrame
print(f"Combined DataFrame shape: {combined_df.shape}")
print(combined_df.head())
```

```python
combined_df
```

```python
combined_df = combined_df[['name', 'main_category', 'sub_category', 'ratings', 'no_of_ratings', 'discount_price', 'actual_price']]
```

```python
combined_df_no_na = combined_df.dropna()
```
